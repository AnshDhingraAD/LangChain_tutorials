**Custom Tools**


Method - 1 : using @tool

```python
from langchain_core.tools import tool
```

```python
@tool
def multiply(a:int , b:int)->int:
  """Multipling 2 numbers"""
  return a*b
```

```python
result=multiply.invoke({'a':2,'b':3})
```

```python
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

Method 2 : using StructuredTool

```python
from langchain_core.tools import StructuredTool
from pydantic import Field,BaseModel
```

```python
class MultiplyInput(BaseModel):
  a:int = Field(required=True , description='First number to multiply')
  b:int = Field(required=True , description='Second number to multiply')
```

```python
def multiply_func(a:int , b:int)->int:
  return a*b
```

```python
multiply_tool=StructuredTool.from_function(
    func=multiply_func,
    name='multiply',
    description='Multiplying 2 numbers',
    args_schema=MultiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

## Method 3 - Using BaseTool Class


```python
from langchain.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):
  a:int=Field(required=True , description='First number to multiply')
  b:int=Field(required=True , description='Second number to multiply')
```

```python
class MultiplyTool(BaseTool):
  name : str = 'Multiply'
  description : str = 'Multiplying 2 numbers'
  args_schema: Type[BaseModel] = MultiplyInput

  def _run(self , a:int , b:int)->int:
    return a*b
```

```python
multiply_tool=MultiplyTool()
```

```python
result=multiply_tool.invoke({'a':3 , 'b':2})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

**Toolkit**

```python
from langchain_core.tools import tool

@tool
def add(a:int , b:int)->int:
  """Add 2 numbers"""
  return a+b

@tool
def multiply(a:int , b:int)->int:
  """Multiply 2 numbers"""
  return a*b
```

```python
class MathToolKit:
  def get_tools(self):
    return [add, multiply]
```

```python
toolkit=MathToolKit()
tools=toolkit.get_tools()

for tool in tools:
  print(tool.name , '->' , tool.description)
```
